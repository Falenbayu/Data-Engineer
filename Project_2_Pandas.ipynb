```python
# Transforming - part 1 - pd.to_datetime()
contoh_list = ['a','b','c','d']
# contoh_datetime1 = ['2023-12-11', '2023-12-31', '2024-01-01', '2024-01-02']
contoh_datetime1 = ['11/12/23', '31/12/23', '01/01/24', '02/01/24']   # penulisan dengan format '%d/%m/%y'
contoh_dict1 = {'order_date': contoh_datetime1, 'type': contoh_list}

df = pd.DataFrame(contoh_dict1)
print('Dataset awal:\n', df)
print('Tipe data awal:\n', df.dtypes)

df['order_date'] = pd.to_datetime(df['order_date'], format='%d/%m/%y')   # tambahkan argumen format='%d/%m/%y'
print('\nDataset sesudah transformasi:\n', df)
print('Tipe data setelah transformasi:\n', df.dtypes)
```

```python
# Transforming - part 2 - pd.to_numeric()
contoh_list = ['a','b','c','c']
contoh_num = ['2', ' 3', '-6', '8']
contoh_dict1 = {'value': contoh_num, 'kategori': contoh_list}

df = pd.DataFrame(contoh_dict1)
print('Dataset awal:\n', df)
print('Tipe data awal:\n', df.dtypes)

df['value'] = pd.to_numeric(df['value'], downcast='float')   # tambahkan argumen downcast='float'
print('\nDataset sesudah transformasi:\n', df)
print('Tipe data setelah transformasi:\n', df.dtypes)
```

```python
# Transforming - part 2 - .astype()
contoh_list = ['a','b','c','c']
contoh_num = ['2', ' 3', '-6', '8']
contoh_dict1 = {'value': contoh_num, 'kategori': contoh_list}

df = pd.DataFrame(contoh_dict1)
print('Tipe data awal:\n', df.dtypes)

df['value'] = df['value'].astype('int32')                     # casting per kolom
df2 = df.astype({'value': 'int32', 'kategori': 'category'})   # casting beberapa kolom
print('\nTipe data setelah transformasi (df):\n', df.dtypes)
print('\nTipe data setelah transformasi (df2):\n', df2.dtypes)
```

```python
# Transforming - part 3 - .applymap() and .apply()
df = pd.DataFrame([[1, 2.12], [3.356, 4.567]])
print('Dataset awal:\n', df)

df1 = df.applymap(lambda x: len(str(x)))        # apply fungsi secara element-wise
df2 = df.iloc[:, 0].map(lambda x: len(str(x)))  # apply fungsi secara element-wise
df3 = df.apply(np.sum, axis=1)                  # apply fungsi di sepanjang axis kolom
df4 = df.apply(np.sum, axis=0)                  # apply fungsi di sepanjang axis baris

print('\nDataset hasil .applymap():\n', df1)
print('\nDataset hasil .map():\n', df2)
print('\nDataset hasil .apply() axis kolom:\n', df3)
print('\nDataset hasil .apply() axis baris:\n', df4)
```

```python
# Inspeksi Missing Value
# Baca file
df=pd.read_csv("https://storage.googleapis.com/dqlab-dataset/datacovid19.csv")
# Cetak info dari df
print(df.info())
# Cetak jumlah missing value di setiap kolom
mv = df.isna().sum()
# df.isna().sum()
print("\nJumlah missing value per kolom:\n", mv)
```

```python
# Treatment Missing Value - part 2 - .dropna()
df = pd.DataFrame({"name": ['Alfred', 'Batman', 'Catwoman'],
                   "toy": [np.nan, 'Batmobile', 'Bullwhip'],
                   "born": [pd.NaT, pd.Timestamp("1940-04-25"),pd.NaT],
                   "dummy": [pd.NA, pd.NA, pd.NA]})
print('Dataset awal:\n', df)

df0_any = df.dropna(axis=0, how='any')    # drop setiap row yang memiliki setidaknya 1 value NA
df1_any = df.dropna(axis=1, how='any')    # drop setiap kolom yang memiliki setidaknya 1 value NA
df0_all = df.dropna(axis=0, how='all')    # drop setiap row yang semua value-nya NA
df1_all = df.dropna(axis=1, how='all')    # drop setiap kolom yang semua value-nya NA

print('\nDataset hasil .dropna(axis=0, how="any"):\n', df0_any)
print('\nDataset hasil .dropna(axis=1, how="any"):\n', df1_any)
print('\nDataset hasil .dropna(axis=0, how="all"):\n', df0_all)
print('\nDataset hasil .dropna(axis=1, how="all"):\n', df1_all)
```

```python
# Treatment Missing Value - part 3 - .fillna()
df = pd.DataFrame([[np.nan, 2, np.nan, 0],
                   [3, 4, np.nan, 1],
                   [np.nan, np.nan, np.nan, 5],
                   [np.nan, 3, np.nan, 4]],
                  columns=list("ABCD"))
print("Dataset awal:")
print(df)
print("\nFill NA dengan 0 : ")
print(df.fillna(0))
print("\nFill NA dengan propagasi \nke depan (method='ffill') : ")
print(df.fillna(method="ffill"))
print("\nReplace NA di kolom 'A','B','C','D' \nmasing-masing dengan 0,1,2,3 : ")
values = {"A": 0, "B": 1, "C": 2, "D": 3}
print(df.fillna(value=values))
print("\nReplace NA dengan nilai \nmean/median/max per kolom:")
values = {"A": df["A"].mean(), "B": df["B"].median(), "C": df["D"].max()}
print(df.fillna(value=values))
```

```python
# Reshaping dataset - .pivot() dan .pivot_table()
import pandas as pd
# Dataframe
data = pd.DataFrame({
  'kelas': 6*['A'] + 6*['B'],
  'murid': 2*['A1'] + 2*['A2'] + 2*['A3'] + 2*['B1'] + 2*['B2'] + 2*['B3'],
  'pelajaran': 6*['math','english'],
  'nilai': [90,60,70,85,50,60,100,40,95,80,60,45]
}, columns=['kelas','murid','pelajaran','nilai'])
print("Dataset awal:")
print(data)

# Pivot dengan index=['kelas','murid'] agar row-index terjamin unique
hasil_pivot = data.pivot(index=['kelas','murid'], columns='pelajaran', values='nilai')
print('\nHasil Pivot dengan index=["kelas","murid"]:')
print(hasil_pivot)

# # Creating pivot and assign pivot_tab dengan menggunakan keyword aggfunc='mean'
pivot_tab_mean = data.pivot_table(index='kelas', columns='pelajaran', values='nilai', aggfunc='mean')
print('Creating pivot table -- aggfunc mean:\n', pivot_tab_mean)

# # Creating pivot and assign pivot_tab dengan menggunakan keyword aggfunc='median'
# pivot_tab_median = data.pivot_table(index='kelas', columns='pelajaran', values='nilai', aggfunc='median')
# print('Creating pivot table -- aggfunc median:\n', pivot_tab_median)
```

```python
# Aggregation & GroupBy - .groupby()
df = pd.DataFrame(
    {
        "A": ["foo", "bar", "foo", "bar", "foo", "bar", "foo", "foo"],
        "B": ["one", "one", "two", "three", "two", "two", "one", "three"],
        "C": np.random.randn(8),
        "D": np.random.randn(8),
    }
)
grouped = df.groupby('A')
for name, group in grouped:
  print(name)
  print(group)
```

```python
# Aggregation & GroupBy - .groupby().<agg_function>
animals = pd.DataFrame(
    {
        "kind": ["cat", "dog", "cat", "dog"],
        "height": [9.1, 6.0, 9.5, 34.0],
        "weight": [7.9, 7.5, 9.9, 198.0],
    }
)
print("Dataset awal:")
print(animals)
# Menggunakan fungsi agregasi setelah groupby
print("\ngroupby dan agregasi sum:")
print( animals.groupby("kind").sum() )
print("\ngroupby dan agregasi mean/average:")
print( animals.groupby("kind").mean() )
# Menggunakan sekumpulan fungsi agregasi setelah groupby
print("\ngroupby dan agregasi min, median, standard deviation:")
print( animals.groupby("kind").agg(['min','median', 'std']) )
```
